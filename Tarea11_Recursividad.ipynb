```
 class Stack:
    def __init__( self ):
        self.__data = list()
 
    def is_empty( self ):
        return len( self.__data ) == 0
 
    def lenght( self ):
        return len( self.__data )
 
    def pop( self ):
        if self.is_empty():
            print("PILA VACIA")
            return
        else:
            return self.__data.pop()
 
    def push( self, value):
        self.__data.append( value )
 
    def peek( self ):
        return self.__data[len( self.__data)-1]
 
    def to_string( self ):
        for item in self.__data[::-1]:
            print("---")
            print(f"|{item}|") 
 
def cuentaRegresiva(num): 
    if num <= 0:
        return num
    else:
        print (num)
    return(cuentaRegresiva(num - 1))
 
def Main_cuentaRegresiva():
    val = 10
    print (f"Cuenta regresiva empezando en : {val}\n")
    print(cuentaRegresiva(val))
 
def mitadPila( pila, val ):
    top = pila.pop()    
    while pila.lenght() != val + 1:
        mitadPila(pila, val)
        pila.push(top)
        return
    else: 
        print (f"El elemento sacado es : {pila.pop()}")
        pila.push(top)
        return
 
def pilaElem ():
    pila = Stack()
    print ("\nPila:")
    pila.push('a')
    pila.push('b')
    pila.push('c')
    pila.push('d')
    pila.push('e')
    pila.push('f')
    pila.push('g')
    pila.push('h')
    pila.push('i') 
    
    return pila 
    
def Main_pila( ):
    pila = pilaElem()
    pila.to_string()
    
    middle = (pila.lenght()/2)  
    if middle == 0.5 or middle <= 1:
        print ("\nNo se pueden sacar mitad de uno o cero elementos")
        return 
    if (middle % 2) == 0:
        print ("\nNumero par de elementos en la pila")
        return
    else:
        middle = int(middle)
 
    print("\nBuscando elemento con recursividad\n")
    mitadPila(pila, middle)
    print("\nPila después de sacar el elemento medio:\n")
    pila.to_string()
 
Main_cuentaRegresiva()
Main_pila()
```
